```python
import sys
import json
import csv

json_file_path_1 = 'train_v1.1.json'
#json_file_path_1 = 'dev_v2.1.json'
json_file_path_2 = 'dev_v1.1.json'

csv_file_path = 'dev_train_marco.csv'


fp1 = open(json_file_path_1, 'r')

json_value = fp1.read()
tweets = []
for line in open(json_file_path_1, 'r'):
    tweets.append(json.loads(line))

fp2 = open(json_file_path_2, 'r')

json_value = fp2.read()
for line in open(json_file_path_2, 'r'):
    tweets.append(json.loads(line))

    
raw_data=tweets
data_to_be_processed = raw_data
processed_data = []
header = []
        

def to_string(s):
    try:
        return str(s)
    except:
        return s.encode('utf-8')


def reduce_item(key, value):
    global reduced_item
    
    #Reduction Condition 1
    if type(value) is list:
        i=0
        for sub_item in value:
            reduce_item(key+'_'+to_string(i), sub_item)
            i=i+1

    #Reduction Condition 2
    elif type(value) is dict:
        sub_keys = value.keys()
        for sub_key in sub_keys:
            reduce_item(key+'_'+to_string(sub_key), value[sub_key])
    
    #Base Condition
    else:
        reduced_item[to_string(key)] = to_string(value)


for item in data_to_be_processed:
    reduced_item = {}
    reduce_item('data',item)
    header += reduced_item.keys()
    processed_data.append(reduced_item)


header = list(set(header))
header.sort()

with open(csv_file_path, 'wb+') as f:
    writer = csv.DictWriter(f, header, quoting=csv.QUOTE_ALL)
    writer.writeheader()
    
    for row in processed_data:
        writer.writerow(row)

print ("Just completed writing csv file with %d columns" % len(header))


```
